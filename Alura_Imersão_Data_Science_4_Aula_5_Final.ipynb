```python

from sklearn.tree import DecisionTreeRegressor


# Create a random dataset

X = dados_vendas_censo[['V005']]
Y = dados_vendas_censo[["Valor_Anuncio_Numeric"]]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25,random_state = 50)

# Fit regression model
regr_1 = DecisionTreeRegressor(max_depth=5)
regr_2 = DecisionTreeRegressor(max_depth=10)
regr_1.fit(X_train, Y_train)
regr_2.fit(X, Y)

# Predict

y_1 = regr_1.predict(X)


# Plot the results
plt.figure()
plt.scatter(X, Y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X, y_1, color="cornflowerblue", label="max_depth=2", linewidth=0.5)
plt.xlabel("Dados")
plt.ylabel("Valor")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

```python
dados_vendas_censo['Decision_Tree_Valor'] = y_1


dados_vendas_censo['Percentual_Decision_Tree'] = dados_vendas_censo['Decision_Tree_Valor']/dados_vendas_censo['Valor_Anuncio_Numeric']*100

dados_vendas_censo[['Valor_Anuncio_Numeric','Valor_Predicao','Percentual_Predicao','Decision_Tree_Valor','Percentual_Decision_Tree']]

```
