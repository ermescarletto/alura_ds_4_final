```python
bairros = dados.Bairro.unique()
lista_medias = []

for i in bairros:
  verifica_imoveis_no_bairro = dados['Bairro'] == i
  imoveis_no_bairro = dados[verifica_imoveis_no_bairro]
  media_metragem_bairro = imoveis_no_bairro['Metragem'].mean()
  lista_medias.append([i,round(media_metragem_bairro,2),imoveis_no_bairro.size])
```

```python
medias = pd.DataFrame(lista_medias,columns=['Bairro','Média','Size'])
print(medias['Média'].max())
print(medias['Média'].min())
medias.sort_values('Size',ascending=False).head(20).plot.bar(x='Bairro')

```
